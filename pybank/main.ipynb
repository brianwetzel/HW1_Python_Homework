```python
# Import path from pathlib
from pathlib import Path
import csv

# Define the path to our data
relative_file_path = Path("budget_data.csv")

# INITIALIZE VARIABLES: Metric Variables
month_counter = 0
profit_loss = 0  
total_profit_loss = 0
monthly_delta = 0
total_delta = 0
max_position = 0
min_position = 0
minimum = 0
maximum = 0

# INITIALIZE VARIABLES: Lists
months_list = []
profit_loss_list = []
delta_list = []
max_list = []
min_list = []

# OPEN DATA: open file based on its path, read it, and store it in a variable called budget_data
with open(relative_file_path, "r") as budget_data:
    # Pass in the csv file to the csv.reader, tell it "," seperates data, and save as object csv_reader
    csv_reader = csv.reader(budget_data, delimiter=",")    
    # create the variable csv_header to move past the header so all code below inlcludes only data (not the header)
    csv_header = next(csv_reader)

    
    # SORT DATA: use "for loop" to assisgn data to lists, and start first calculations 
    for row in csv_reader:
        # Set headers as variables for better readability, and convert strings to integers for calculations.
        month = row[0]
        profit_loss = int(row[1])
        # Adds data to lists
        months_list.append(month)
        profit_loss_list.append(profit_loss)
        # Count months: a for loop that counts the rows by starting at 0 and adding 1 for each row.
        month_counter = month_counter + 1 
        # Total_profit_loss is calculated. The += adds two values together and assigns the result to a variable.  
        total_profit_loss += profit_loss
    
    # CREATE DELTA LIST: calculate the delta in profit/loss between adjacent months, and add the result to the delta_list
    for i in range(1, len(profit_loss_list)):
        delta_list.append(profit_loss_list[i] - profit_loss_list[i-1])
    
    # SUM DELTA LIST
    for delta in delta_list:
        total_delta += delta
        
    # CALCULATE AVERAGE DELTA: by dividing total delta by the total number entries in our delta list (months-1)
    average_total_delta = (total_delta / (month_counter - 1))
    
    # CALCULATE AVERAGE PROFIT/LOSS: by dividing total profit/loss by the total number of months
    average_profit_loss = (total_profit_loss / month_counter)
    
    # MAX
    # use the max() functions to find the max amount in the delta_list, and return what position its in
    maximum = max(delta_list)
    max_position = delta_list.index(maximum)
    # use the position returned above to look up the corresponding date
    max_date = months_list[int(max_position)+1]
    
    # MIN
    # Use the min() functions to find the max amount in the delta_list, and return what position its in
    minimum = min(delta_list)
    min_position = delta_list.index(minimum)
    # Use the position returned above to look up the corresponding date
    min_date = months_list[int(min_position)+1]
   

with open('financial_analysis.txt','w') as file:
    file.write(f"Financial Analysis" '\n')
    file.write(f"------------------" '\n')
    file.write(f"The total number of months in this data set is {month_counter}." '\n')
    file.write(f"The total profit/loss is ${total_profit_loss:,.2f}." '\n')
    file.write(f"The average monthly change in Profit/Losses: ${average_total_delta:,.2f}" '\n')
    file.write(f"The greatest monthly increase in profits: {max_date}, ${maximum:,.2f}" '\n')
    file.write(f"The greatest monthly decrease in losses {min_date}, ${minimum:,.2f}" '\n')
```
